```python
# 7) What are the top 5 acquiring institutions?
banks['Acquiring Institution'].value_counts().head(5)
```

```python
# 8) How many banks has the State Bank of Texas acquired?

(banks["Acquiring Institution"] == "State Bank of Texas").sum()
```

```python
# how many of them were actually in Texas
banks[(banks["Acquiring Institution"] == "State Bank of Texas") & (banks["ST"] == "TX")].shape[0]
```

```python
# 9) What is the most common city in California for a bank to fail in?

banks[banks["ST"] == "CA"]["City"].value_counts().idxmax()
```
