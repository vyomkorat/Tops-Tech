```python
category_subscribers_mean=data.groupby("category")["subscribers"].mean().round(2)
category_subscribers_mean.sort_values(ascending=False)
```

##### This analysis shows that **'Shows'** and **'Trailers'** categories have the highest average subscribers, both over **39 million**. Categories like **Travel & Events** and **Autos & Vehicles** have the lowest averages. This means viewer interest is higher in entertainment-heavy content.


```python
category_subscribers_mean.sort_values(ascending=False,inplace=True)
category_subscribers_mean.plot(kind="bar",color='red')
plt.title('Mean Subscribers By Category')
plt.xlabel('Category')
plt.ylabel('Mean Subscribers')
plt.xticks(rotation=90)
plt.show()
```

##### This Bar chart analysis shows that **'Shows'** and **'Trailers'** categories have the highest average subscribers, both over **39 million**. Categories like **Travel & Events** and **Autos & Vehicles** have the lowest averages. This means viewer interest is higher in entertainment-heavy content.

```python
def millions(val):
    if np.isnan(val):
        return ""
    elif val >= 1_000_000:
        return f"{val / 1_000_000:.1f} M"
    elif val >= 1_000:
        return f"{val / 1_000:.0f} K"
    else:
        return str(int(val))


pivot = data.groupby(['Country', 'created_year'])['highest_monthly_earnings'].mean().unstack()
annot = pivot.apply(np.vectorize(millions))


plt.figure(figsize=(20, 12))  
sns.heatmap(pivot, annot=annot, fmt='', cmap='YlGnBu', linewidths=0.5, annot_kws={"size": 9})
plt.title('Avg. Highest Monthly Earnings by Year and Country', fontsize=18)
plt.xlabel('Created Year', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(rotation=0, fontsize=11)
plt.tight_layout()
plt.show()
```

##### The heatmap shows the average highest monthly YouTube earnings across countries and years. It highlights that countries like South Korea is the highest earnings in specific years (up to $6.8M), while many others had low or no data. This helps identify top-earning countries and peak periods in YouTube monetization trends.

```python
channel_created_subscribers_sum=data.groupby(['channel_type','created_year'])['subscribers_for_last_30_days'].sum().unstack()
```

```python
channel_created_subscribers_sum
```

```python
def millions(val):
    if np.isnan(val):
        return ""
    elif val >= 1_000_000:
        return f"{val / 1_000_000:.0f} M"
    elif val >= 1_000:
        return f"{val / 1_000:.0f} K"
    else:
        return str(int(val))


annot = channel_created_subscribers_sum.apply(np.vectorize(millions))


plt.figure(figsize=(14, 8))
sns.heatmap(channel_created_subscribers_sum, cmap='YlGnBu', annot=annot, fmt='', linewidths=0.5)
plt.title('Sum Of  Subscribers in Last 30 Days by Channel Type and Year (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Channel Type')
plt.xticks(rotation=45)
plt.show()
```

##### This heatmap shows that the "Entertainment" and "People" categories experienced the highest subscriber growth in recent years, especially around 2015–2020. It also highlights which content types were most popular in specific years.

```python
data.groupby('created_month')['subscribers_for_last_30_days'].mean().plot(kind='line',marker='o')
plt.title('Average subscribers for last 30 days by month')
plt.xlabel("Month")
plt.ylabel("Average Subscribers for last 30 days")
plt.show()
```

##### From this chart, we see that the average number of subscribers gained in the last 30 days varies by month. December, January, and June show higher subscriber gains, while May has the lowest. This suggests seasonal trends may influence subscriber growth.
