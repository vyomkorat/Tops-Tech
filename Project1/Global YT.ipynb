```python
category_subscribers_mean=data.groupby("category")["subscribers"].mean().round(2)
category_subscribers_mean.sort_values(ascending=False)
```

##### This analysis shows that **'Shows'** and **'Trailers'** categories have the highest average subscribers, both over **39 million**. Categories like **Travel & Events** and **Autos & Vehicles** have the lowest averages. This means viewer interest is higher in entertainment-heavy content.


```python
category_subscribers_mean.sort_values(ascending=False,inplace=True)
category_subscribers_mean.plot(kind="bar",color='red')
plt.title('Mean Subscribers By Category')
plt.xlabel('Category')
plt.ylabel('Mean Subscribers')
plt.xticks(rotation=90)
plt.show()
```

##### This Bar chart analysis shows that **'Shows'** and **'Trailers'** categories have the highest average subscribers, both over **39 million**. Categories like **Travel & Events** and **Autos & Vehicles** have the lowest averages. This means viewer interest is higher in entertainment-heavy content.


```python
channel_types_count = data['channel_type'].value_counts()
channel_types_count
```

##### This analysis shows that Entertainment(304) and Music(216) are the most common channel types,While Less common types like Autos(3), Animals(3), and Nonprofit(2) very few channels 

```python
channel_types_count.head().plot(kind='pie',autopct="%1.1f%%")
plt.title("Channel Tye Disrtibution")
plt.show()
```

##### This pie chart shows that most YouTube channels belong to the **Entertainment** (39.5%) and **Music** (28.1%) types. These are the most popular among creators, while **Comedy**, **Games**, and **People** channels make up smaller portions.


```python
plt.figure(figsize=(10, 6))
plt.scatter(data["subscribers"], data["video views"], alpha=0.5)
plt.title("Subscribers vs Video Views")
plt.xlabel("Subscribers")
plt.ylabel("Video Views")
plt.tight_layout()
plt.show()
```

##### This scatter plot shows that **more subscribers usually lead to more video views**, but not always. Some channels with fewer subscribers have high views, and some with many subscribers have fewer views, suggesting other factors like content quality or virality also matter.


```python
country_subscribers_total = data.groupby('Country')['subscribers'].sum().sort_values(ascending=False)
country_subscribers_total
```

##### This analysis shows that the **United States** and **India** have the highest total YouTube subscribers, followed by **Brazil** and the **UK**. This means YouTube is most popular or has the most active creators in these countries.


```python
country_subscribers_total.head(10).plot(kind='bar')
plt.title('Total subscribers by country')
plt.xlabel('country')
plt.ylabel('Total subscribers')
plt.xticks(rotation=90)
plt.show()
```

##### This Bar chart analysis shows that the **United States** and **India** have the highest total YouTube subscribers, followed by **Brazil** and the **UK**. This means YouTube is most popular or has the most active creators in these countries.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="channel_type", y="highest_monthly_earnings", data=data, palette='Set2',hue='channel_type',legend=False)
plt.title("Distribution of Highest Monthly Earning By Channel Type")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

##### 

* **Animal** channels have the highest and most spread out monthly earnings compared to other channel types.
* Most other categories like **Entertainment**, **Comedy**, and **News** also earn well, but with many smaller outliers.
* **Tech** and **Nonprofit** channels generally have lower earnings and less variation.


```python
data.groupby(['category', 'Country'])['subscribers'].sum().unstack().plot(kind="bar", stacked=True)
plt.title('Subscribers Breakdown by Category and Country')
plt.xlabel("Category")
plt.ylabel('Total Subscribers')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(rotation=90)
plt.show()
```

##### 

* **Music**, **Entertainment**, and **Education** are the top categories in terms of total YouTube subscribers.
* Countries like **India**, **Vietnam**, and **United States** contribute significantly to subscriber counts in many categories.
* Some categories like **Nonprofits & Activism** and **Trailers** have relatively low subscriber engagement across all countries.


```python
plt.scatter(data["Gross tertiary education enrollment (%)"],data["subscribers"],alpha=0.5)
plt.title('scatter plot of subscribers vs gross terrtiary eduction enrollment')
plt.xlabel("Gross tertiary eduction enrollment")
plt.ylabel("subscribers")
plt.show()
```

##### Some countries with low education enrollment still have many subscribers, while some with high education enrollment have fewer.
##### higher education doesn’t always mean more subscribers, and many other factors could be influencing YouTube popularity.

```python
data.groupby("channel_type")['Unemployment rate'].mean().head(5).sort_values(ascending=False).plot(kind='bar')
plt.title('Average umployment rate by channel type')
plt.xlabel('channel type')
plt.ylabel('Average unmployment rate')
plt.show()
```

##### This bar chart shows the average unemployment rate across different YouTube channel types. It reveals that channels related to "Animals" have the highest average unemployment rate, while "Autos" channels have the lowest. This might suggest that creators in the animal-related content category come from regions with higher unemployment.


```python
grouped = data.groupby(['created_year', 'channel_type'])['subscribers'].mean()
top_10_years = grouped.groupby(level=0).mean().sort_values(ascending=False).head(10).index
filtered = grouped.loc[top_10_years]
filtered.unstack().plot(kind='bar', width=1, figsize=(12, 6))

plt.title("Average Subscribers by Created Year and Channel Type (Top 10 Years)")
plt.xlabel('Created Year')
plt.ylabel('Average Subscribers')
plt.legend(title='Channel Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

##### This analysis shows the top 10 years where YouTube channels gained the highest average subscribers. For example, channels created around 2005–2007 and 2013 attracted the most subscribers on average. Entertainment and Education channels were among the top-performing types during these peak years.


```python
data.groupby(['created_year', 'created_month'])['subscribers'].sum().unstack().head().plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Subscribers Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Total Subscribers')
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

##### This analysis shows subscriber growth across different years and months. The year 2006 saw the highest total subscriber growth, especially in March and September. It also indicates that subscriber accumulation is influenced by both year and month of channel creation.


```python
data.groupby(['created_year', 'Country'])['highest_monthly_earnings'].sum().unstack().head().plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Trend of Highest Yearly Earnings by Created Year and Country')
plt.xlabel('Created Year')
plt.ylabel('Total Highest Yearly Earnings')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

##### This analysis shows that some countries like Vietnam and Brazil experienced a sharp rise in the highest yearly earnings around 2006. It also reveals how earnings trends differ by country

```python
category_highest_monthly_earnings_sum=data.groupby('category')['highest_monthly_earnings'].sum()
category_highest_monthly_earnings_sum.sort_values(ascending=False,inplace=True)
```

```python
category_highest_monthly_earnings_sum
```

##### This analysis shows that the Entertainment and Music categories generate the highest total monthly earnings, far ahead of others. Categories like Travel & Events and Nonprofits earn the least, indicating a strong income gap between popular and niche content types.


```python
category_highest_monthly_earnings_sum.head().plot(kind='pie',autopct='%1.1f%%')
plt.title('Percentage of highest monthly earning by category')
plt.ylabel('')
plt.show()
```

##### This pie chart shows that the Entertainment category contributes the most to the highest monthly earnings (36.3%), followed by Music and People & Blogs. These three categories alone make up over 80% of top earnings, highlighting their strong audience appeal.


```python
channel_created_subscribers_mean=data.groupby(['channel_type','created_year'])['subscribers_for_last_30_days'].mean().unstack()
```

```python
channel_created_subscribers_mean
```

```python
plt.figure(figsize=(14, 8))
sns.heatmap(channel_created_subscribers_mean, cmap='YlGnBu', annot=True, fmt=".0f", linewidths=0.5)

plt.title('Mean Subscribers in Last 30 Days by Channel Type and Year (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Channel Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 
* Entertainment, Comedy, and People channels consistently attracted high subscriber growth across many years, especially post-2013.
* Some categories like Tech and Education show strong spikes in specific years (e.g., 2020), likely due to increased digital engagement.
* Other categories such as Animals, Nonprofit, and Autos show low or inconsistent subscriber growth over the years.

```python
data.groupby('created_month')['subscribers_for_last_30_days'].mean().plot(kind='line',marker='o')
plt.title('Average subscribers for last 30 days by month')
plt.xlabel("Month")
plt.ylabel("Average Subscribers for last 30 days")
plt.show()
```

##### From this chart, we see that the average number of subscribers gained in the last 30 days varies by month. December, January, and June show higher subscriber gains, while May has the lowest. This suggests seasonal trends may influence subscriber growth.
